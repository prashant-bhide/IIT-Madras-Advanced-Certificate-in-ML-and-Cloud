## Deployment Testing

```python
# Importing the required libraries

import requests
from pandas import json_normalize
```

### Deployment API testing on local flask app instance

```python
# We will invoke the 'predict_campaigns' API from Flask application deployed on local instance
# and display the Testing dataset device_id to Campaigns mapping for 5 random samples

host_url = 'http://127.0.0.1:5000/predict_campaigns/'
num_samples = 5
url = host_url + str(num_samples)
response = requests.get(url)
dict = response.json()
df = json_normalize(dict)
df.head(num_samples)
```

### Deployment API testing on Docker Container deployed on EC2 instance

```python
# We will invoke the 'predict_campaigns' API from Docker Container deployed on EC2 instance
# and display the Testing dataset device_id to Campaigns mapping for 50 random samples

host_url = 'http://ec2-3-80-102-136.compute-1.amazonaws.com:5000/predict_campaigns/'
num_samples = 50
url = host_url + str(num_samples)
response = requests.get(url)
dict = response.json()
df = json_normalize(dict)
df.head(num_samples)
```
